## END TO END DATA PIPELINE PROJECT
SCRAPPING DATA(HEADLINES AND CORRESPONDING HYPERLINKS) FROM AN ONLINE NEWS WEBSITE (DAWN.COM)

```python
import pandas as pd
import requests

from bs4 import BeautifulSoup as BS
import numpy as np
print("Libraries Successfully Imported!")
```

```python
url = "https://www.dawn.com/"
extracting_data = requests.get(url).text
wiki_data = BS(extracting_data, 'html.parser')
```

```python
Headlines = []
links = []
story = []
data = []
i = 0
for news in wiki_data.findAll('h2' , {'class':'story__title'}):
    datum={}
    Headlines.append(news.text.strip('\n'))
    links.append(news.a['href'])
    datum['Headlines'] = Headlines[i]
    datum['Links'] = links[i]
    data.append(datum)
    i+=1


```
